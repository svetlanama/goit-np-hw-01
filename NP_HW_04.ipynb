5. Стандартизуйте дані:

```python
# 5. Стандартизація даних

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Завантаження даних
data = load_breast_cancer()

# Створення DataFrame
df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target

# Стандартизуємо лише числові ознаки (без 'target')
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[data.feature_names])

# Створюємо новий DataFrame зі стандартизованими даними
df_scaled = pd.DataFrame(scaled_data, columns=data.feature_names)
df_scaled["target"] = df["target"]

# Виведемо перші 5 рядків стандартизованих даних
print("Перші 5 рядків стандартизованих даних:")
print(df_scaled.head())

# Перевіримо середнє та стандартне відхилення
print("\nСереднє після стандартизації (очікується близько 0):")
print(df_scaled[data.feature_names].mean().round(2))

print("\nСтандартне відхилення після стандартизації (очікується близько 1):")
print(df_scaled[data.feature_names].std().round(2))
```

## 🔹 2. Distance Matrices (Cityblock / Cosine / Euclidean)

| Metric | Description | Observation |
|---------|--------------|-------------|
| **Cityblock (Manhattan)** | Measures the sum of absolute coordinate differences (L₁ norm). Robust to outliers. | Produces evenly distributed distances. Samples from the same class are closer, though overall contrast is moderate. |
| **Euclidean** | Measures straight-line geometric distance in feature space. | Similar to Cityblock but shows slightly stronger contrast between near and distant points. |
| **Cosine** | Measures angle (direction) between feature vectors, ignoring magnitude. | Shows the **strongest contrast** — samples within the same class are closest, while samples between classes are furthest apart. |

---

## 🔹 3. PCA 3D Visualization
- The **3D PCA projection** confirms two main clusters corresponding to the two tumor classes.  
- **Benign (blue)** and **malignant (red)** samples occupy distinct regions in the PCA space.  
- Minimal overlap suggests that the first three principal components already capture most of the class-related variance.

---

## 🔹 4. Overall Insights
- All metrics effectively reflect class separability after standardization.  
- **Cosine distance** emphasizes orientation and pattern similarity, yielding the **clearest separation**.  
- **Euclidean** and **Cityblock** are stable, general-purpose metrics suitable for biomedical data with normalized scales.
